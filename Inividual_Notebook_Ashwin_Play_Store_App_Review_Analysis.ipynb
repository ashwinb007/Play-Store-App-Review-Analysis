#<u>Data cleaning

```python
# here we will start cleaning the data of Playstore by removing all the duplicates
play_store_df.drop_duplicates(inplace = True)
```

```python
# we will check whether there is any null values in the data.
play_store_df.isnull().sum()
```

```python
#we will check the null values for android version
play_store_df[play_store_df['Android Ver'].isnull()]
```

```python
# here we got to know that 1st two records are null values but the in third all the columns are misplaced. so we will drop it.
play_store_df.drop(10472,inplace = True)
```

```python
#now we will change the null value of Android Version.
play_store_df['Android Ver'].fillna(play_store_df['Android Ver'].mode,inplace = True)
```

```python
#we will check the null values for current version
play_store_df[play_store_df['Current Ver'].isnull()]
```

```python
#now we will change the null value of Current Version.
play_store_df['Current Ver'].fillna(play_store_df['Current Ver'].mode,inplace = True)
```

```python
#same way we will do for Type column
play_store_df['Type'].fillna(play_store_df['Type'].mode,inplace = True)
```

```python
#next we will fill the null values of ratings by median
play_store_df['Rating'].fillna(play_store_df['Rating'].median(),inplace = True)
```

```python
#now we have filled all the null values as you can see below data.
play_store_df.isnull().sum()
```

```python
play_store_df.info()
```

# now we need to change the following into numeric values.
  
  Reviews\
  Installs\
  Price
  

```python
play_store_df.head()
```

```python
play_store_df = play_store_df.astype({'Reviews' : int})
```

```python
play_store_df['Installs'].unique()
```

```python
# we will remove '+' and ',' so that we can convert into numerical value
play_store_df['Installs'] = play_store_df['Installs'].str.replace('+','').str.replace(',','')
play_store_df = play_store_df.astype({'Installs' : int})
```

```python
play_store_df['Price'].unique()
```

```python
# we will remove '$' so that we can convert into numerical value
play_store_df['Price'] = play_store_df['Price'].str.replace('$','')
play_store_df = play_store_df.astype({'Price' : float})
```

```python
# u can see here we got the numerical values.
play_store_df.info()
```

```python
# this the the final dataframe of playstore where we will do the analysis.
play_store_df
```

```python
# now we will start cleaning the data of User reviews.
# we will remove all the duplicates
user_review_df.drop_duplicates(inplace = True)
```

```python
user_review_df.info()
```

```python
user_review_df.isnull().sum()
```

```python
user_review_df['Sentiment'].unique()
```

```python
# we are going to remove all the rows whose sentiments are null.
user_review_df = user_review_df[~user_review_df['Sentiment'].isna()]
```

```python
user_review_df.isnull().sum()
```

```python
user_review_df.fillna(0,inplace = True)
```

```python
# this the the final dataframe of playstore where we will do the analysis.
user_review_df
```

# Top Content Rating Values

```python
Crv_graph = sns.countplot(play_store_df['Content Rating'])
plt.rcParams['figure.figsize'] = (15, 10)
plt.title('Number of Apps based on Content Rating')
plt.show()
```

**INFERENCE:** As we can see in the plot, the most number of apps by Content rating is for everyone. Everyone can use most of the apps. And apps which is based on least content Ratings are of ' Adults only 18+' and 'Unrated' as there are only fews apps apps available of these two content ratings.

# Distribution of Ratings

```python
Dor_graph = sns.countplot(play_store_df['Rating'])
plt.rcParams['figure.figsize'] = (15, 10)
plt.title('Count of Ratings')
plt.show()
```

**INFERENCE:** We found out that most of the apps ratings are between 4.2 and 4.6. Maximum numbers of apps rated is 4.5. There are very fews apps which is rated less than 3.6.

#Revenue generated by Top Categories

```python
#Revenue = Price * Installs
play_store_df['Revenue'] = play_store_df['Price'] * play_store_df['Installs']
```

```python
play_store_df.sort_values(by="Revenue",ascending=False).head()
```

```python
play_store_df[["App","Revenue","Category"]]
```

```python
revenue_category = play_store_df.groupby("Category",as_index=False)["Revenue"].sum().sort_values(by="Revenue",ascending=False).head(15)
revenue_category
```

```python
sns.barplot(x=revenue_category["Revenue"],y=revenue_category["Category"],data=revenue_category)
plt.title("Revenue generated by Top Category",fontsize=20)
plt.show()
```

**INFERENCE:** From this graph, we can understand that the Category who generated the highest revenue is 'Family' followed by 'Lifestyle' and 'Game'.

# App Price across Categories

```python
papc = sns.stripplot(x = play_store_df['Price'], y = play_store_df['Category'], linewidth=1 , size=8)
papc.set_title('Trend of App Price across categories',size=30)
plt.show()
```

**Inference:** We can understand from these graphs that the top prices are of 'Family', 'Lifestyle', and 'Finance' categories.

# Correlation of both the Datas

```python
# now we will merge both data for more analysis.
whole_df = play_store_df.merge(user_review_df, on="App")
whole_df.head()
```

```python
whole_df.info()
```

```python
sns.heatmap(whole_df.corr(),annot = True,cmap='Greens')
plt.title("Correlation on merge datas",fontsize=20)
plt.show()
```

**Inference:** The above plot shows the correlation of both the datas i.e Play Store data and User Review Data. We found that the most of the correlations between variables are positive and only some are negative.

# Age vs Sentiments

```python
avs = sns.countplot(whole_df["Content Rating"],hue=whole_df['Sentiment'],palette="Set3")
plt.title("Age vs Sentiment",size=15)
for p in avs.patches:
    avs.annotate(p.get_height(),(p.get_x() + p.get_width()/2,p.get_height()),ha='center',size= 12)
plt.show()
```

**Inference:** From this graph, we can understand that most sentiments are from 'everyone'. which is positive. followed by negative. The second most sentiments are from 'Teen'.
